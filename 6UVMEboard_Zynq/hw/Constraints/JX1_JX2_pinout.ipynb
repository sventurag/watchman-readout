```python
import pandas as pd
import numpy as np
from datetime import date
```

# MicroZed_JX_pinout_tables_jxX.csv files based on [MicroZed JX1 JX2 Pinout Tables](http://zedboard.org/support/documentation/1519) and previously filled manually using 6U VME schematics ports

```python
jx1=pd.read_csv('MicroZed_JX_pinout_tables_jx1.csv', sep="\t") 
```

```python
jx2=pd.read_csv('MicroZed_JX_pinout_tables_jx2.csv', sep="\t")
```

```python
jx1_vme = jx2["VME"].values #Values were interchanged in first manual draft
```

```python
jx2_vme = jx1["VME"].values #Values were interchanged in first manual draft
jx2.head(50)
```

# JX2

```python
jx2.drop(columns="VME", inplace=True)
```

```python
jx2["VME"] = jx2_vme
jx2= jx2.reset_index(drop=True)
jx2.set_index("JX2 Pin", inplace=True)

#jx2.loc[10,'VME']="PG_1V8_ZED"
```

```python
jx2.head(50)
```

## Corrections to original file

```python
jx2.loc[8,'VME']="MIO9"
#jx2.loc[6,'VME']="SPI_MISO"
jx2.loc[7,'VME']="MIO0"
jx2.loc[11,'VME']="PG_ZED"
jx2.loc[94,'VME']="B.DO_9"
```

```python
jx2["VME"] = jx2["VME"].str.replace('.','_') 
jx2["VME"] = jx2["VME"].str.upper()
```

## Making constraints commands

```python
jx2.loc[ (jx2['Zynq Pin'].notnull()) & (jx2['VME'].notnull()),   'PACKAGE_PIN'] = 'set_property PACKAGE_PIN ' + jx2['Zynq Pin'].map(str) + ' [get_ports '+ jx2['VME'].map(str) + ']' 
jx2.loc[ (jx2['Zynq Pin'].notnull()) & (jx2['VME'].notnull()),   'IOSTANDARD'] = 'set_property IOSTANDARD LVCMOS25' +' [get_ports '+ jx2['VME'].map(str) + ']' 
jx2['pin_commented'] = '# '+ jx2.index.map(str)
jx2.head(10)
```

```python
jx2.head(50)
```

```python
jx1["VME"] = jx1_vme
jx1= jx1.reset_index(drop=True)
jx1.set_index("JX1 Pin", inplace=True)
```

## Making constraints commands

```python
jx1.loc[ (jx1['Zynq Pin'].notnull()) & (jx1['VME'].notnull()),   'PACKAGE_PIN'] = 'set_property PACKAGE_PIN ' + jx1['Zynq Pin'].map(str) + ' [get_ports '+ jx1['VME'].map(str) + ']' 
jx1.loc[ (jx1['Zynq Pin'].notnull()) & (jx1['VME'].notnull()),   'IOSTANDARD'] = 'set_property IOSTANDARD LVCMOS25' +' [get_ports '+ jx1['VME'].map(str) + ']' 
jx1['pin_commented'] = '# '+ jx1.index.map(str)
jx1.head(51)
```

## JX1 AND JX2 ARE SEPARATED BY ODD AND EVEN PIN NUMBERS to easily check pin out

```python
jx1_odd= jx1.iloc[::2]
jx1_even= jx1.iloc[1::2]
```

```python
jx2_odd= jx2.iloc[::2]
jx2_even= jx2.iloc[1::2]
```

## Write constraints file

```python
def writeTitle(text):
    file= open('6UVME_TARGETC_BOARD_constraints.csv','a')
    file.write('{} \n'.format(text))
    file.close()
    
```

```python
file= open('6UVME_TARGETC_BOARD_constraints.csv','w')
file.write('# PINOUT Constraint File  \n')
file.write('# 6U VME WATCHMAN BOARD  \n')
file.write('# Authors: Blaine Furman, Salvador Ventura \n')
file.write('# {} \n'.format(date.today()))
file.write('set_property PULLUP true [get_ports SDA] \n')
file.write('set_property PULLUP true [get_ports SCL]  \n')
file.write('#### #### #### #### #### #### #### #### #### #### #### #### \n')
file.write('# PACKAGE_PIN \n')
file.write('#### #### #### #### #### #### #### #### #### #### #### #### \n')
file.write('# JX1 odd pins \n')
file.close()
jx1_odd.loc[ jx1['PACKAGE_PIN'].notnull(), ['PACKAGE_PIN', 'pin_commented'] ].reset_index(drop=True).to_csv('6UVME_TARGETC_BOARD_constraints.csv', doublequote=False, index=False, header=False, mode='a', sep='\t')
writeTitle('# JX1 even pins')
jx1_even.loc[ jx1['PACKAGE_PIN'].notnull(), ['PACKAGE_PIN', 'pin_commented'] ].reset_index(drop=True).to_csv('6UVME_TARGETC_BOARD_constraints.csv', doublequote=False, index=False, header=False, mode='a', sep='\t')
writeTitle('# JX2 odd pins')
jx2_odd.loc[ jx2['PACKAGE_PIN'].notnull(), ['PACKAGE_PIN', 'pin_commented'] ].reset_index(drop=True).to_csv('6UVME_TARGETC_BOARD_constraints.csv', doublequote=False, index=False, mode='a', header=False, sep='\t')
writeTitle('# JX2 even pins')
jx2_even.loc[ jx2['PACKAGE_PIN'].notnull(), ['PACKAGE_PIN', 'pin_commented']].reset_index(drop=True).to_csv('6UVME_TARGETC_BOARD_constraints.csv', doublequote=False, index=False, mode='a', header=False, sep='\t')

file= open('6UVME_TARGETC_BOARD_constraints.csv','a')
file.write('#### #### #### #### #### #### #### #### #### #### #### #### \n')
file.write('# IOSTANDARD \n')
file.write('#### #### #### #### #### #### #### #### #### #### #### #### \n')
file.write('# JX1 odd pins \n')
file.close()
jx1_odd.loc[ jx1['IOSTANDARD'].notnull(),  ['IOSTANDARD', 'pin_commented'] ].reset_index(drop=True).to_csv('6UVME_TARGETC_BOARD_constraints.csv', doublequote=False, index=False, mode='a', header=False, sep='\t')
writeTitle('# JX1 even pins')
jx1_even.loc[ jx1['IOSTANDARD'].notnull(), ['IOSTANDARD', 'pin_commented'] ].reset_index(drop=True).to_csv('6UVME_TARGETC_BOARD_constraints.csv', doublequote=False, index=False, mode='a', header=False, sep='\t')
writeTitle('# JX2 odd pins')
jx2_odd.loc[ jx2['IOSTANDARD'].notnull(), ['IOSTANDARD', 'pin_commented'] ].reset_index(drop=True).to_csv('6UVME_TARGETC_BOARD_constraints.csv', doublequote=False, index=False, mode='a', header=False, sep='\t')
writeTitle('# JX2 even pins')
jx2_even.loc[ jx2['IOSTANDARD'].notnull(), ['IOSTANDARD', 'pin_commented'] ].reset_index(drop=True).to_csv('6UVME_TARGETC_BOARD_constraints.csv', doublequote=False,index=False, mode='a', header=False, sep='\t')
```
